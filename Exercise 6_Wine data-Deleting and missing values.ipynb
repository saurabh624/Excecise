### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.at[:2,'alcohol'] = np.nan
wine.head()
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.loc[[3,4],'Magnisum']=np.nan
wine.head()
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
#wine[['alcohol']].fillna(10)
wine['alcohol'].replace(np.nan,10,inplace = True)
wine.head()
```

```python
#wine[['Magnisum']].fillna(100)
wine['Magnisum'].replace(np.nan,100,inplace = True) ## its mean that inplace of np.nan replace with 100
wine.head()
```
